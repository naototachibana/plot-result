```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# def plot_logfile (filepath):
#     log_file = filepath
#     df = pd.read_json(log_file)
#     med_rmse = df['validation/main/root_mean_sqr_error'].median()*2
#     max_epoch = df['epoch'].max()
#     print (med_rmse)
#     print (max_epoch)
#     df.plot(y=['main/root_mean_sqr_error', 'validation/main/root_mean_sqr_error'],
#             xlim=[0, max_epoch],
#             ylim=[0,med_rmse],
#             fontsize=20,
#             figsize=(15,10),
#             grid=True)
# #     plt.scatter(X, train_RMSE,
# #                     color = 'red',
# #                     s = 2,
# #                     marker = 'o',
# #                     label='train RMSE')
# #     plt.scatter(X, validation_RMSE,
# #                     color = 'blue',
# #                     s = 2,
# #                     marker = 'o',
# #                     label='validation RMSE') 
    
def plot_logfile (filepath):
    log_file = filepath
    df = pd.read_json(log_file)
    df['main/root_mean_sqr_error']= df['main/root_mean_sqr_error']
    df['validation/main/root_mean_sqr_error'] = df['validation/main/root_mean_sqr_error']
    med_RMSE = df['validation/main/root_mean_sqr_error'].median()
    min_RMSE_train = df['main/root_mean_sqr_error'].min()
    min_RMSE_valid = df['validation/main/root_mean_sqr_error'].min()
    max_epoch = df['epoch'].max()
    print ("min RMSE of train dataset : " + str(min_RMSE_train))
    print ("min RMSE of valid dataset : " + str(min_RMSE_valid))
    print ("number of epochs : " + str(max_epoch))
    X = df['epoch'].to_numpy()
    
    train_RMSE = df['main/root_mean_sqr_error'].to_numpy()
    validation_RMSE = df['validation/main/root_mean_sqr_error'].to_numpy()


    plt.plot(X, train_RMSE, color = 'red', label='train RMSE')
    plt.plot(X, validation_RMSE, color = 'blue', label='validation RMSE') 
    

    plt.xlabel('epochs')
    plt.ylabel('RMSE')
    
    plt.xlim(0, max_epoch)
    plt.ylim(0, med_RMSE*1.5)
    plt.grid()
    plt.legend()
```

```python
plot_logfile('../eval_total_energy_kcal_mol_lowest_nfp_0_2000epoch/log_report')
```

```python
plot_logfile('../eval_HOMO_kcal_mol_lowest_nfp_0_2000epoch/log_report')
```
